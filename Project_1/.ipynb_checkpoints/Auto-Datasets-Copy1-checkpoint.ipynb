## 4. Feature Engineering

```python
df.info()
```

```python
df["normalized-losses"].median()
```

```python
df["bore"].fillna(df["bore"].median())
```

```python
# df["stroke"].fillna(df["stroke"].mode()[0]).astype(float).astype(int)
# df["horsepower"].fillna("111").astype(int)
```

```python
data = df.copy()
```

```python
# changing data type of feature as per business understanding
data["normalized-losses"] = data["normalized-losses"].fillna(data["normalized-losses"].median()).astype(int)
data["bore"]              = data["bore"].fillna(data["bore"].median()).astype(float)
data["stroke"]            = data["stroke"].fillna(data["stroke"].mode()[0]).astype(float)
data["horsepower"]        = data["horsepower"].fillna("111").astype(int)
data["peak-rpm"]          = data["peak-rpm"].fillna("5000").astype(int)
data["num-of-doors"]      = data["num-of-doors"].fillna(data["num-of-doors"].mode()[0])
```

```python
data["num-of-doors"].mode()[0]
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
# missing rows for price column
data[data["price"].isna()]
# np.select()
```

```python
lst = [(data["make"] == "audi") & (data["price"].isna()),
      (data["make"] == "isuzu") & (data["price"].isna()),
      (data["make"] == "porsche") & (data["price"].isna())]

cond = [(data[data["make"]=="audi"]["price"].sort_values(ascending=False).reset_index().drop("index",axis=1).iloc[0][0]),
       (data[data["make"]=="isuzu"]["price"].sort_values(ascending=False).reset_index().drop("index",axis=1).iloc[0][0]),
      (data[data["make"]=="porsche"]["price"].sort_values(ascending=False).reset_index().drop("index",axis=1).iloc[0][0]),
       ]
data["price"] = np.select(lst,cond,default =data["price"])
```

## 6) Model Training

## 6.1 spliting data sets

```python
x = df2.drop("price",axis=1) 
y = df2["price"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

## 6.2 Model fitting

```python
lr_model = LinearRegression()
lr_model.fit(x_train,y_train)
```

## 7 Model Evaluation

```python
# Training
y_pred = lr_model.predict(x_train)
mse = mean_squared_error(y_train,y_pred)
print("MSE",mse)
mae  = mean_absolute_error(y_train,y_pred)
print("MAE",mae)
r2 = r2_score(y_train,y_pred)
print("R2",r2)
```

```python
# Testing
y_pred = lr_model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
print("MSE",mse)
mae  = mean_absolute_error(y_test,y_pred)
print("MAE",mae)
r2 = r2_score(y_test,y_pred)
print("R2",r2)
```
